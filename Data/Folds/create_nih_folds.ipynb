```python
import pandas as pd
import os
```

```python
root = '/media/hdd/Kaggle/Pneumothorax/Data/Folds'
folds = [os.path.join(root, x) for x in os.listdir(root) if x.endswith('.csv')]
holdout = folds[-1]
folds = folds[:-1]
folds
```

```python
nih_csv_path = '/media/nvme/Datasets/Pneumothorax/NIH/nih_ptx.csv'
nih_df = pd.read_csv(nih_csv_path) 
print(nih_df.shape)
nih_df.tail()
```

## Add EncodedPixels

```python
masks = pd.read_csv('/media/hdd/Kaggle/Pneumothorax/Output/seg_ens_nih.csv')
nih_df = nih_df.set_index('ImageId')
nih_df.loc[masks['ImageId'], 'EncodedPixels'] = masks['EncodedPixels'].values
nih_df.tail()
```

## verify contamination and create folds

```python
nih_df.loc[nih_df['SIIM_ID'].isna(), 'SIIM_ID'] = nih_df.loc[nih_df['SIIM_ID'].isna()].index.values
nih_df = nih_df.set_index('SIIM_ID')
nih_df['Set'] = 'train'

holdout_df = pd.read_csv(holdout)
holdout_df = holdout_df[holdout_df['EncodedPixels'] != '-1']
print('{} ptx images in nihset'.format(len(nih_df)))
for i, fold in enumerate(folds):
    df = pd.read_csv(fold)
    df = df[df['EncodedPixels'] != '-1']
    df = pd.concat([df, holdout_df])
    df = df.set_index('ImageId')
    print('{} ptx images in fold{} + holdout'.format(len(df.index.unique()), i))
    
    intersect = list(set(nih_df.index.values).intersection(set(df.index.unique().values)))
    new_df = nih_df.drop(intersect)
    new_df = pd.concat([new_df, df], sort=False)
    print('{} ptx images in new_set. {} train, {} val'.format(len(new_df.index.unique()),
                                                          len(new_df[new_df['Set'] == 'train'].index.unique()),
                                                          len(new_df[new_df['Set'] == 'val'].index.unique())))
    new_df = new_df.reset_index().rename(columns={'index':'ImageId'})

    new_df.to_csv(fold.replace('.csv', '_nih.csv'), index=False)
    
```
